## 🧍 3. What is Postural Analysis? <a name='3'></a>

**Postural analysis** is the evaluation of body alignment in static and dynamic positions to identify deviations from optimal posture.

### Common Postural Deviations
| Deviation | Region | Description |
|---|---|---|
| Forward Head Posture | Cervical | Head displaced anterior to gravity line |
| Kyphosis | Thoracic | Excessive rounding of upper back |
| Lordosis | Lumbar | Excessive inward curve of lower back |
| Scoliosis | Spinal | Lateral curvature of the spine |
| Flat Foot | Foot | Collapsed medial arch |
| Genu Valgum | Knee | Knock-knee deformity |

## 🕴️ 6. How AI Detects Posture — MediaPipe Pose Estimation <a name='6'></a>

**MediaPipe** by Google detects **33 body landmarks** in real-time from a camera feed. Key landmarks for postural analysis:
- Ear, Shoulder, Hip, Knee, Ankle → detect sagittal alignment
- Left/Right Shoulder, Hip → detect frontal tilt

### Key Angle Calculations
| Angle | Landmarks Used | Normal Range |
|---|---|---|
| Craniovertebral Angle | Ear–Shoulder | > 50° |
| Trunk Inclination | Shoulder–Hip–Vertical | 0–5° |
| Knee Flexion | Hip–Knee–Ankle | 170–180° |
| Shoulder Symmetry | L.Shoulder–R.Shoulder | < 2° difference |

```python
# ── Simulate MediaPipe Landmark Detection ──
# (Works without a camera — uses synthetic landmark data)

def calculate_angle(A, B, C):
    """
    Calculate the angle at point B formed by vectors BA and BC.
    Points are (x, y) tuples.
    """
    A, B, C = np.array(A), np.array(B), np.array(C)
    BA = A - B
    BC = C - B
    cos_angle = np.dot(BA, BC) / (np.linalg.norm(BA) * np.linalg.norm(BC) + 1e-6)
    angle = np.degrees(np.arccos(np.clip(cos_angle, -1, 1)))
    return round(angle, 1)

# Synthetic landmark coordinates (x, y) for a person with forward head posture
# Using a coordinate system where (0,0) is top-left
landmarks = {
    'ear':       (0.55, 0.12),   # slightly forward
    'shoulder':  (0.50, 0.25),
    'hip':       (0.50, 0.55),
    'knee':      (0.50, 0.75),
    'ankle':     (0.50, 0.95),
    'elbow':     (0.35, 0.42),
    'wrist':     (0.30, 0.58),
}

# Calculate clinical angles
cv_angle     = calculate_angle(landmarks['ear'],      landmarks['shoulder'], (landmarks['shoulder'][0], 0))
trunk_angle  = calculate_angle(landmarks['shoulder'], landmarks['hip'],      (landmarks['hip'][0], 0))
knee_angle   = calculate_angle(landmarks['hip'],      landmarks['knee'],     landmarks['ankle'])

print('='*55)
print('  🦴  AI POSTURAL ANALYSIS — ANGLE REPORT')
print('='*55)
print(f'  Craniovertebral Angle : {cv_angle}°   (Normal: >50°)')
print(f'  Trunk Inclination     : {trunk_angle}°  (Normal: 0–5°)')
print(f'  Knee Extension Angle  : {knee_angle}° (Normal: 170–180°)')
print('='*55)
print()

# Flags
if cv_angle < 50:
    print(f'  ⚠️  Forward Head Posture detected (CV angle: {cv_angle}° < 50°)')
else:
    print(f'  ✅  Craniovertebral angle within normal limits')

if trunk_angle > 5:
    print(f'  ⚠️  Trunk tilt detected ({trunk_angle}° > 5°)')
else:
    print(f'  ✅  Trunk alignment normal')

if knee_angle < 170:
    print(f'  ⚠️  Knee flexion deformity detected ({knee_angle}° < 170°)')
else:
    print(f'  ✅  Knee angle within normal limits')
```

```python
# ── Visualise the Synthetic Pose with Landmarks ──
fig, axes = plt.subplots(1, 2, figsize=(12, 7))

def draw_pose(ax, lm, title, color):
    # Skeleton connections
    connections = [
        ('ear', 'shoulder'), ('shoulder', 'hip'),
        ('hip', 'knee'), ('knee', 'ankle'),
        ('shoulder', 'elbow'), ('elbow', 'wrist')
    ]
    ax.set_facecolor('#EEF2FF')
    for p1, p2 in connections:
        x_vals = [lm[p1][0], lm[p2][0]]
        y_vals = [lm[p1][1], lm[p2][1]]
        ax.plot(x_vals, y_vals, '-', color=color, linewidth=4, solid_capstyle='round')
    for name, (x, y) in lm.items():
        ax.scatter(x, y, s=150, color='white', edgecolors=color, linewidth=3, zorder=5)
        ax.text(x + 0.03, y, name.capitalize(), fontsize=9, color='#333')
    ax.set_xlim(0, 1)
    ax.set_ylim(1, 0)  # flip y-axis (image coords)
    ax.set_title(title, fontweight='bold', fontsize=12)
    ax.axis('off')

# Normal posture
normal_lm = {
    'ear':      (0.50, 0.12), 'shoulder': (0.50, 0.25),
    'hip':      (0.50, 0.55), 'knee':     (0.50, 0.75),
    'ankle':    (0.50, 0.95), 'elbow':    (0.35, 0.42),
    'wrist':    (0.30, 0.58),
}
draw_pose(axes[0], normal_lm, '✅ Normal Posture', '#1565C0')
axes[0].axvline(0.50, color='green', linestyle='--', alpha=0.4, label='Gravity Line')
axes[0].legend(loc='lower center')

# Abnormal posture (forward head)
draw_pose(axes[1], landmarks, '⚠️ Forward Head Posture Detected', '#EF5350')
axes[1].axvline(0.50, color='green', linestyle='--', alpha=0.4, label='Gravity Line')
axes[1].legend(loc='lower center')

plt.suptitle('MediaPipe Pose Landmark Detection — Physiotherapy View',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('pose_landmarks.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 🤖 7. Common Postural Deviations — AI Classification Model <a name='7'></a>

We will train a **Random Forest Classifier** on synthetic postural angle data to classify:
- Normal Posture
- Forward Head Posture
- Kyphosis
- Lordosis
- Scoliosis

```python
# ── Generate Synthetic Postural Dataset ──
np.random.seed(42)
n = 300  # samples per class

def make_samples(n, cv, trunk_sagittal, trunk_frontal, lumbar, shoulder_sym, label):
    """
    cv              : Craniovertebral angle
    trunk_sagittal  : Thoracic kyphosis angle
    trunk_frontal   : Lateral trunk tilt
    lumbar          : Lumbar lordosis angle
    shoulder_sym    : Shoulder height difference (cm)
    """
    data = {
        'cv_angle':         np.random.normal(cv, 4, n),
        'trunk_sagittal':   np.random.normal(trunk_sagittal, 3, n),
        'trunk_frontal':    np.random.normal(trunk_frontal, 2, n),
        'lumbar_angle':     np.random.normal(lumbar, 4, n),
        'shoulder_sym':     np.random.normal(shoulder_sym, 0.5, n),
        'label': label
    }
    return pd.DataFrame(data)

#               cv   t_sag  t_fro  lumb  sym
normal  = make_samples(n, 55,   5,    1,   45,  0.2,  'Normal')
fhp     = make_samples(n, 38,   8,    2,   48,  0.4,  'Forward Head')
kyphosis= make_samples(n, 50,  40,    2,   35,  0.5,  'Kyphosis')
lordosis= make_samples(n, 52,   6,    2,   65,  0.4,  'Lordosis')
scolio  = make_samples(n, 51,  10,   15,   47,  2.5,  'Scoliosis')

df = pd.concat([normal, fhp, kyphosis, lordosis, scolio], ignore_index=True)

print('Dataset shape:', df.shape)
print()
print(df.groupby('label').size().to_string())
print()
df.head()
```

```python
# ── Train the Random Forest Classifier ──
features = ['cv_angle', 'trunk_sagittal', 'trunk_frontal', 'lumbar_angle', 'shoulder_sym']
X = df[features]
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

clf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f'  ✅ Model Accuracy: {acc*100:.1f}%')
print()
print(classification_report(y_test, y_pred))
```

```python
# ── Confusion Matrix & Feature Importance ──
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=clf.classes_, yticklabels=clf.classes_,
            ax=axes[0], linewidths=0.5)
axes[0].set_title('🎯 Confusion Matrix — Postural Classification', fontweight='bold')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

# Feature importance
importances = clf.feature_importances_
feat_df = pd.Series(importances, index=features).sort_values(ascending=True)
colors_feat = ['#1565C0', '#1976D2', '#42A5F5', '#90CAF9', '#BBDEFB']
feat_df.plot(kind='barh', ax=axes[1], color=colors_feat, edgecolor='white')
axes[1].set_title('🔑 Feature Importance in Classification', fontweight='bold')
axes[1].set_xlabel('Importance Score')
for i, (val, name) in enumerate(zip(feat_df.values, feat_df.index)):
    axes[1].text(val + 0.002, i, f'{val:.3f}', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('model_results.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 📱 9. APECS App Simulation <a name='9'></a>

**APECS** = AI Posture Evaluation & Correction System  
A mobile AI app that:
1. Captures posture from smartphone camera
2. Detects 33 body landmarks
3. Calculates joint angles
4. Generates clinical report + exercise plan
5. Syncs with cloud for progress tracking

```python
# ── APECS App — Full Clinical Report Simulation ──
import datetime

def generate_apecs_report(patient_name, age, angles):
    """
    Simulate an APECS App clinical posture report.
    angles: dict with keys: cv, thoracic, lateral, lumbar, shoulder_diff
    """
    cv         = angles['cv']
    thoracic   = angles['thoracic']
    lateral    = angles['lateral']
    lumbar     = angles['lumbar']
    sh_diff    = angles['shoulder_diff']

    findings = []
    score = 100

    if cv < 50:
        findings.append(f'  ⚠️  Forward Head Posture — CV Angle: {cv}° (Normal: >50°)')
        score -= 20
    else:
        findings.append(f'  ✅  Craniovertebral angle normal ({cv}°)')

    if thoracic > 40:
        findings.append(f'  ⚠️  Hyperkyphosis — Thoracic Angle: {thoracic}° (Normal: <40°)')
        score -= 20
    else:
        findings.append(f'  ✅  Thoracic kyphosis within range ({thoracic}°)')

    if lateral > 5:
        findings.append(f'  ⚠️  Lateral Trunk Deviation: {lateral}° (Normal: <5°) — Possible Scoliosis')
        score -= 20
    else:
        findings.append(f'  ✅  Lateral alignment normal ({lateral}°)')

    if lumbar > 60:
        findings.append(f'  ⚠️  Hyperlordosis — Lumbar Angle: {lumbar}° (Normal: 40–60°)')
        score -= 20
    elif lumbar < 40:
        findings.append(f'  ⚠️  Flat Back — Lumbar Angle: {lumbar}° (Normal: 40–60°)')
        score -= 10
    else:
        findings.append(f'  ✅  Lumbar curve within normal limits ({lumbar}°)')

    if sh_diff > 1.5:
        findings.append(f'  ⚠️  Shoulder Asymmetry: {sh_diff} cm (Normal: <1.5 cm)')
        score -= 10
    else:
        findings.append(f'  ✅  Shoulder symmetry acceptable ({sh_diff} cm)')

    grade = 'Excellent' if score >= 90 else 'Good' if score >= 75 else 'Fair' if score >= 55 else 'Poor'
    date = datetime.date.today().strftime('%d %B %Y')

    print('='*60)
    print('         📱  APECS — AI POSTURE REPORT')
    print('='*60)
    print(f'  Patient : {patient_name}         Age: {age}')
    print(f'  Date    : {date}')
    print(f'  Posture Score: {score}/100  [{grade}]')
    print('-'*60)
    print('  FINDINGS:')
    for f in findings:
        print(f)
    print('-'*60)
    print('  RECOMMENDATIONS:')
    if score < 75:
        print('  📌 Refer to physiotherapist for detailed assessment')
        print('  📌 Begin structured corrective exercise programme')
        print('  📌 Postural awareness training recommended')
    else:
        print('  📌 Continue general fitness and postural hygiene')
        print('  📌 Ergonomic review of workstation advised')
    print('='*60)
    return score

# Run the simulation
score = generate_apecs_report(
    patient_name='John Doe',
    age=35,
    angles={'cv': 42, 'thoracic': 44, 'lateral': 3, 'lumbar': 55, 'shoulder_diff': 0.8}
)
```

## 🏋️ 10. Corrective Exercise Planning with AI <a name='10'></a>

Based on the postural diagnosis, AI generates a **3-phase exercise programme**:

| Phase | Goal | Duration |
|---|---|---|
| 1 — Stretch | Release tight muscles | Weeks 1–2 |
| 2 — Strengthen | Activate weak muscles | Weeks 3–6 |
| 3 — Stabilise | Neuromuscular control | Weeks 7–12 |

```python
# ── AI Exercise Prescription Engine ──

EXERCISE_DATABASE = {
    'Forward Head': {
        'stretch':    ['Chin Tucks', 'Suboccipital Release', 'Pec Stretch'],
        'strengthen': ['Deep Neck Flexor Training', 'Scapular Retraction', 'Wall Angels'],
        'stabilise':  ['Head-on-Hand Isometrics', 'Thoracic Extension', 'Postural Mirror Training'],
    },
    'Kyphosis': {
        'stretch':    ['Thoracic Doorway Stretch', 'Foam Roller Extension', 'Chest Opener'],
        'strengthen': ['Prone Row', 'Face Pulls', 'Thoracic Extension on Roller'],
        'stabilise':  ['Bird-Dog', 'Dead Bug', 'Swiss Ball Back Extension'],
    },
    'Lordosis': {
        'stretch':    ['Hip Flexor Stretch', 'Rectus Femoris Stretch', 'Iliopsoas Release'],
        'strengthen': ['Abdominal Hollowing', 'Glute Bridge', 'Dead Bug'],
        'stabilise':  ['Plank', 'McGill Big 3', 'Pelvic Stability Drill'],
    },
    'Scoliosis': {
        'stretch':    ['Side Bending Stretch', 'Concave Side Release', 'Cat-Camel'],
        'strengthen': ['Schroth Method Breathing', 'Asymmetric Side Plank', 'Concave Lat Pull'],
        'stabilise':  ['3D Breathing', 'Mirror Therapy', 'Proprioceptive Balance Training'],
    },
    'Normal': {
        'stretch':    ['General Mobility Routine'],
        'strengthen': ['Core Strengthening Programme'],
        'stabilise':  ['Balance & Proprioception Training'],
    }
}

def prescribe_exercises(diagnosis):
    db = EXERCISE_DATABASE.get(diagnosis, EXERCISE_DATABASE['Normal'])
    print('='*60)
    print(f'  💊  AI EXERCISE PRESCRIPTION — {diagnosis.upper()}')
    print('='*60)
    for phase, label, weeks in [
        ('stretch',    '🔵 Phase 1 — STRETCH  (Weeks 1–2)',  'x 30 sec, 3 reps, daily'),
        ('strengthen', '🟠 Phase 2 — STRENGTHEN (Weeks 3–6)', 'x 3 sets of 12, 3×/week'),
        ('stabilise',  '🟢 Phase 3 — STABILISE (Weeks 7–12)', 'x 3 sets, 3×/week'),
    ]:
        print(f'\n  {label}')
        print(f'  Frequency: {weeks}')
        for i, ex in enumerate(db[phase], 1):
            print(f'    {i}. {ex}')
    print('='*60)

# Try different diagnoses
prescribe_exercises('Forward Head')
print()
prescribe_exercises('Kyphosis')
```
